```python
a = []
for i in range(10):
  a.append(list(map(int, input().split())))

x, y = 1, 1
a[x][y] = 9

# 
for i in range(100):
  if x == 9 and y==9:
    break
  if a[x][y+1] == 1 and a[x+1][y] ==1:
      break
  if a[x][y+1] == 2:
    a[x][y+1] = 9
    break
  elif a[x][y+1] == 0:
    y += 1
    a[x][y] = 9
  elif a[x+1][y] == 2:
    a[x+1][y] = 9
    break
  elif a[x+1][y] == 0:
    x += 1
    a[x][y] = 9



for i in range(len(a)):
  for j in range(len(a)):
    print(a[i][j], end=' ')
  print()
```
